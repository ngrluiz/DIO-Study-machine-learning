# **Descrição do Desafio**

Seguindo o exemplo do algoritmo de binarização apresentado em nossa última aula, realize a implementação em Python para transformar uma imagem colorida para níveis de cinza (0 a 255) e para binarizada (0 e 255), preto e branco.  

```python
#fazendo o import das bibliotecas / importing the library

import cv2
import matplotlib.pyplot as plt
```

```python
#lendo a imagem / reading the images
img = cv2.imread('Lenna.png')
```

```python
#convertendo para tons de cinza / Converting to Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#binarizando (preto e branco)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
#Exibindo resultados / printing the results
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title('Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('OFF')

plt.subplot(1,3,2)
plt.title('Grayscale')
plt.imshow(gray, cmap='gray')
plt.axis('OFF')

plt.subplot(1,3,1)
plt.title('Preto e Branco (binario)')
plt.imshow(binary, cmap='gray')
plt.axis('OFF')

plt.show()
```
